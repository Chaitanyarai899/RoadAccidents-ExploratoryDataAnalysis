### Introduction

India is a country having a high usage of vehicles. The vehicle consumption has drastically increased in the last 40 years from `6 million to 230 million` vehicles. Due to the increasing rate of **9%** vehicles per year, the occurence of road accidents has increased exponentially which in turn has hampered the road security of the people in India. 

This project dwells into greater depths in order to decipher the ambiguity in th e traffic security conditions of the country that are causing the possible accidents.

### Data

The data for the project has been collected from [Open Government Data Portal](https://data.gov.in/dataset-group-name/road-accidents), an authorized portal by the Government of India. The data consists of multiple datasets sprawling over various paramters of judgements like type of vehicles, time of occurence, natural/unnatural reasons of the accident etc.
